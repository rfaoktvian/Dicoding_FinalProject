```python
import kagglehub
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### LOAD DATASET

```python
path = kagglehub.dataset_download("pacificrm/skindiseasedataset")

print("Path to dataset files:", path)
```

```python
import os

for root, dirs, files in os.walk(path):
    print("Folder:", root)
    print("Subfolder:", dirs[:5])
    print("Jumlah file:", len(files))
    break
```

```python
train_dir = os.path.join(path, "SkinDisease", "SkinDisease","train")

classes = os.listdir(train_dir)
print("Jumlah kelas:", len(classes))
print(classes[:10])
```

```python
from collections import Counter

class_counts = {}

for cls in classes:
    class_counts[cls] = len(os.listdir(os.path.join(train_dir, cls)))

print(class_counts)
```

```python
from PIL import Image
import random

sample_class = random.choice(classes)
img_path = os.path.join(train_dir, sample_class, random.choice(os.listdir(os.path.join(train_dir, sample_class))))

img = Image.open(img_path)

plt.imshow(img)
plt.title(sample_class)
plt.axis('off')
```

```python
sizes = []

for cls in classes[:5]:  
    files = os.listdir(os.path.join(train_dir, cls))[:10]
    for f in files:
        img = Image.open(os.path.join(train_dir, cls, f))
        sizes.append(img.size)

print(sizes[:10])
```

Berdasarkan pengecekan ukuran gambar di atas, terdapat inkonsistensi dari tiap tiap gambar yang ada di dataset, dimana dari informasi ini dapat disimpulkan bahwa perlu adanya *RESIZE* gambar untuk menyamakan ukuran standar gambar di dataset
